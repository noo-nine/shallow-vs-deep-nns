## data extraction

### extract data from the image by using 'grid/pixel scanning'

```python
import matplotlib.image as mpimg
import csv
import pandas as pd
```

```python
img = mpimg.imread('Baarle-Nassau_Baarle-Hertog.png')
```

```python
# img in the form of numpy array now
# unpacking
```

```python
height, width, channels = img.shape
```

```python
data = []
```

```python
for y in range(height):
    for x in range(width):
        #rgb format
        r, g, b = img[y, x][:3]
        # yellow vs white
        if r > 0.8 and g > 0.8 and b < 0.3:
            label = 1 # yellow - low blue intensity
        else:
            label = 0
        
        data.append([x, y, label])

# storing in a csv file 
with open('map_data.csv', 'w', newline = '') as f:
    writer = csv.writer(f)
    writer.writerow(["x", "y", "label"])
    writer.writerows(data)
```
